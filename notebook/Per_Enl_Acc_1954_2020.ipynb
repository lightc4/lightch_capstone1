```python
print('Line graph:')
x = [1955, 1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020]
plt.figure(figsize=(10,6))
plt.tick_params(left = False) 
plt.xticks(x, rotation=45)
plt.plot(df['Year'], df['DoDEnl_Per_of_Civ'])
plt.ylim(0, 35)

plt.xlabel('Year')
plt.ylabel('Enlisted Percent (%)')
plt.title('Percent Enlisted Accessions US Armed Forces, Age 18 Only, 1954-2020')
plt.show()
```
