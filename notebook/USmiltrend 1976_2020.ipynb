```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('../data/Mil_app_acc_76_20.csv')
```

```python
df.head()
```

```python
#https://www.geeksforgeeks.org/plot-multiple-lines-in-matplotlib/
#https://www.geeksforgeeks.org/how-to-set-x-axis-values-in-matplotlib-in-python/
#https://www.geeksforgeeks.org/how-to-change-line-color-in-matplotlib/
x = [1976, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020]
plt.figure(figsize=(10,6))
plt.tick_params(left = False) 
plt.plot(df['Fiscal Year(FY)'], df['Applicants'], label = 'Applicants', color = 'blue')
plt.plot(df['Fiscal Year(FY)'], df['Newly Enlisted'], label = 'Enlisted', color = 'red')

plt.ylim(100000, 900000)

plt.title('Number of Applicants vs People Enlisting in the US Armed Forces, 1976-2020')
plt.xticks(x, rotation=45)
plt.legend()
plt.show()
```
